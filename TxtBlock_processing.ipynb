```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
#读取xml文件，获取文本块和章节标题

from xml.dom.minidom import parse
import xml.dom.minidom
import sys
```

```python
defaultencoding = 'utf-8'
if sys.getdefaultencoding() != defaultencoding:
    reload(sys)
    sys.setdefaultencoding(defaultencoding)
```

```python
# 使用minidom解析器打开 XML 文档
DOMTree = xml.dom.minidom.parse("D:\\LWJ_2018\\博士论文\\地质调查事件提取\\code\\xml\\内蒙古自治区15万区域矿产地质调查报告.xml")
collection = DOMTree.documentElement  #获取dom对象元素
```

```python
SectionNames=collection.getElementsByTagName("SectionName") #获取节点列表
for i in range(len(SectionNames)):
    print(SectionNames[i].firstChild.data)   #打印节点数据
```

```python
SectionContents=collection.getElementsByTagName("SectionContent") #获取节点列表
SectionC = []
for i in range(len(SectionContents)):
    print(SectionContents[i].firstChild.data)   #打印节点数据
    SectionC.append(SectionContents[i].firstChild.data)  #SectionC的类型为<class 'list'>
```
