处理pdf转成的txt文件
主要操作：
    若该行文本长度为0则删除该行
    km2替换为平方千米；m替换为米（替换前提为km2或m前为数字）
    

```python
def re_ml(fileLines, fileTxt):
    i = 0
    ml = '...........'
    fileTxt_ml = []
    for i in range(fileLines):
        if ml not in fileTxt[i]:
            fileTxt_ml.append(fileTxt[i])
        i = i + 1
    return fileTxt_ml
```

```python
def readTxt(fileName):
    with open(fileName, 'r',encoding = 'utf-8') as file_to_read:
        fileTxt = [] #fileTxt类型为list
        fileLines = 0

        lines = file_to_read.readlines() # 读取全部内容，并以列表方式返回
        for line in lines:
            if line.isspace() == False:    #删除空格行
                fileTxt.append(line.replace(" ", ""))
                fileLines = fileLines + 1
        fileTxt_ml = re_ml(fileLines, fileTxt)
        return fileTxt_ml
```

```python
def saveTxt(txtList, file):
    txt = ''
    with open(file, "w",encoding='utf-8') as f:
        f.write(txt.join(txtList))
```

```python
#获取文件夹下的txt文件
import os
path = 'pdf'  # pdf文件的路径
save_path = 'txt'
dirs = os.listdir(path)  #调用listdir方法遍历pdf文件夹
for file in dirs:
    if os.path.splitext(file)[1] == '.txt':
        t = os.path.splitext(file)[0]
        #print(t)  #打印所有txt格式的文件名
        fileName = path + '/' + t + ".txt"
        fileTxt_ml = readTxt(fileName)
        save_file = save_path + '/' + t + ".txt"
        saveTxt(fileTxt_ml, save_file)
```
